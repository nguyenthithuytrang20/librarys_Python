**YÊU CẦU:** Vẽ biểu đồ giá gas của Japan, thiết lập các tham số để thu được biểu đồ như sau:
<img src='Pic/line1.jpg' width='800px'>

```python
#Gọi các thư viện sử dụng
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#---------------------------------
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Data/gas_prices.csv')
data
```

```python
#Lấy dữ liệu năm và giá gas USA
x = list(data['Year'])
y = list(data['USA'])
print('Time:',x)
print('Values:',y)
```

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
x_Japan = list(data['Year'])
y_Japan = list(data['Japan'])
print('Time:',x_Japan)
print('Values:',y_Japan)
plt.plot(x_Japan,y_Japan) #Vẽ biểu đồ với các tham số mặc định
# vé đồ thị
plt.figure(figsize = (10,4)) #Thiết lập kích thước biểu đồ

plt.plot(x_Japan,                 #Dữ liệu trục X
         y_Japan,                 #Dữ liệu trục Y
         color='#26734d',         #Màu của đường
         linestyle='-.',    #Kiểu đường
         linewidth=1.0,     #Độ rộng của đường line
         marker='D',        #Kiểu điểm
         markersize = 5)   #Kích thước điểm

#Tiêu đề của đồ thị
plt.title('BIỂU ĐỒ THỂ HIỆN GIÁ GAS CỦA NHẬT GIAI ĐOẠN 1990-2008')
#Nhãn cho trục X
plt.xlabel('Thời gian')
#Nhãn cho trục Y
plt.ylabel('Giá GAS')

#Setup giới hạn cho trục X:
plt.xlim(1988,2008)

#Setup giới hạn cho trục Y:
plt.ylim(0,8)

#Hiển thị lưới:
plt.grid(axis='y',
         c='#cccccc',
         ls='-',
         lw=0.5)

plt.show()
```

**Phương thức legend():** Hiển thị chú thích của biểu đồ
Bao gồm các tham số chính:
1. loc: Xác định vị trí hiển thị của chú thích trong biểu đồ, gồm các tuỳ chọn sau:
* 'best'| 0
* 'upper right' | 1
* 'upper left' | 2
* 'lower left' | 3
* 'lower right' | 4
* 'right' | 5
* 'center left' | 6
* 'center right' | 7
* 'lower center' | 8
* 'upper center' | 9
* 'center' | 10

2. ncol: Số cột của chú thích (số nguyên, mặc định là 1)
3. fontsize: kích thước font chữ trong chú thích
4. labelcolor: Màu chữ trong chú thích (mặc định màu đen)
5. facecolor: Màu nền của ô chú thích (mặc định None)
6. title: Dòng tiêu đề trong chú thích

https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.legend.html

## THỰC HÀNH 3:
---
**YÊU CẦU:** Sử dụng dữ liệu đã có vẽ biểu đồ Pie chart và Bar chart như sau:
<img src='Pic/pie1.jpg' width='900px'>

```python
#Tạo dữ liệu: Số lượng SV Nam - Nữ theo từng khoá.
labels = ['K60','K61','K62','K63','K64','K65']
boys  = [200, 340, 260, 440, 300, 180]
girls = [30,  60,  90,  160, 180, 220]
total = list(np.array(boys) + np.array(girls))

sex =['Nam','Nữ']
sum_boy = sum(boys)
sum_girl = sum(girls)
val = [sum_boy,sum_girl]

# Chuẩn bị dữ liệu:
print(sex)
print(val)
print('số lượng sinh viên theo từng khoa:',total)
```

```python
plt.figure(figsize = (15,5))
# làm nổi bật một phần 
plt.subplot(1,2,1)
e=[0,0]
plt.pie(val,
        labels = sex,
        autopct='%.2f %%',
        pctdistance=0.7,
        startangle=90,
        textprops={'color':'w','fontweight':'bold'},
        wedgeprops=dict(edgecolor='w'),#Đường viền màu trắng
        explode=e)#Làm nổi bật một phần)

plt.title('Biểu đồ tỷ lên SV Nam - Nữ của khoa CNTT', fontdict={'fontname':'Tahoma',
                                                    'fontweight':'bold',
                                                   'fontsize':15})
plt.legend(loc=4)
#--------------------------------------------------------------------
plt.subplot(1,2,2)

bar = plt.bar(labels,total,color='#2929a3', width = 0.7)
bar[3].set_color('#ff9933')

#Tiêu đề của đồ thị
plt.title('Biểu đồ số lương SV theo từng khóa', fontdict={'fontname':'Arial','fontweight':'bold','fontsize':15})
plt.grid(axis='y',ls='--')

plt.show()
```

## THỰC HÀNH 4:

**YÊU CẦU:** Sử dụng dữ liệu đã có vẽ biểu đồ Pie chart và Line chart như sau:

```python
plt.figure(figsize=(14,4))
plt.subplot(1,2,1)
plt.plot(labels,total,'r-s', label='Tổng số sinh viên')
plt.plot(labels,boys,'b--o', label='Sinh viên Nam')
plt.plot(labels,girls,'k--o', label='Sinh viên Nữ')

# Đặt tên biểu đồ
plt.title("Biểu đồ sinh viên Nam Nữ, tổng số SV theo từng Khóa", fontdict={'fontweight':'bold',
                                                                          'fontsize':11})
# Hiển thị lưới

plt.grid(ls='--')
plt.legend(loc=4)
#---------------------------------------------------------------
e = [0,0.08,0,0.18,0.05,0]
plt.subplot(1,2,2)
plt.pie(total,
       labels = labels,
       autopct='%.2f %%',
       pctdistance=0.7,
       startangle=180,
       textprops={'color':'k','fontweight':'bold'},
       wedgeprops=dict(width=0.7,edgecolor='w'),# Đường viền màu trắng
       explode=e)#Làm nổi bật một phần)
plt.title('Biểu đồ tỷ lệ SV của từng khóa',fontdict={'fontweight':'bold',
                                                     'fontsize':12})
plt.show()
```

```python
# Gọi các thư viện sử dụng
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#xử lý lỗi
import warnings
warnings.filterwarnings('ignore')
```

# Yêu cầu 1:
---
Thực hiện thống kê số lượng Nam – Nữ trong tập dữ liệu và trực quan hóa kết quả:

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}
# thống kê số lượng nam, nữ
sl_gender=bmi[['Gender']].value_counts()
gt=['Male','Female']

# Trực quan hóa
plt.figure(figsize = (10,4))
plt.pie(sl_gender,
       labels=gt,
       autopct='%.2f %%',
       textprops={'color':'k','fontweight':'bold'})
plt.title('Thống kê số lượng nam nữ', fontdict={'fontweight':'bold',
                                               'fontsize':15})
plt.show()
```

## Yêu cầu 2:
---
Thực hiện thống kê số lượng theo chỉ số cơ thể (index) và trực quan hóa kết quả.
* 0 - Extremely Weak
* 1 - Weak
* 2 - Normal
* 3 - Overweight
* 4 - Obesity
* 5 - Extreme Obesity

```python
change_index={0:'Extremely Weak',
             1:'Weak',
             2:'Normal',
             3:'Overweight',
             4:'Obesity',
             5:'Extreme Obesity'}
bmi['Index'].replace(change_index,inplace=True)
print(bmi)
```

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY}

# Thống kê số lượng theo chỉ số cơ thể
sl_index=bmi[['Index']].value_counts()
print(sl_index)

# Lấy dữu liệu
#labels=['Extremely Weak','Weak','Normal','Overweight','Obesity','Extreme Obesity']
labels=bmi['Index'].unique()
#Trực quan hóa

plt.figure(figsize=(12,4))# kích thước

plt.subplot(1,2,1)
plt.barh(labels,
         sl_index,# giá trị tương ứng với nhãn
         color='#990000',
         height = 0.6)
plt.xlabel('Giá GAS ($)')


plt.grid(axis='x',ls='--',color='g')# Hiển thị lưới
#----------------------------------------------------------------------------
plt.subplot(1,2,2)
plt.pie(sl_index,
       labels=labels,
       autopct='%.2f %%',
       pctdistance=0.6,
       startangle=180,
       textprops={'color':'k','fontweight':'bold'},#thiết lập label
       labeldistance=1.1, #Khoảng cách từ label tới biểu đồ
       wedgeprops=dict(width=0.7,edgecolor='w'),#Đường viền màu trắng
       rotatelabels=False)#Label có xoay không?
plt.suptitle('THỐNG KÊ DỮU LIỆU THEO TRẠNG THÁI CƠ THỂ TRONG TẬP DỮ LIỆU ')
plt.show()



```

## Yêu cầu 5:
---
Vẽ biểu đồ thể hiện mối tương quan giữa Height và Weight của 500 người--> cho nhận xét?

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# đọc dữ liệu
data_500=pd.read_csv('Data/Data_500.csv', sep=';')
print(data_500)
# Lấy dữ liệu
height=data_500['Height'].values
weight=data_500['Weight'].values

print('Chiều cao:',height)
print('Cân nặng:',weight)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(height,
            weight,
           color='r',
           alpha=0.5)
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Cân nặng (kg)')
plt.title('BIểu ĐỒ THỂ HIỆN MỐI QUAN HỆ GIỮA CHIỀU CAO VÀ CÂN NẶNG CỦA TẬP DỮ LIỆU', fontdict={'fontweight':'bold',
                                                                                               'fontsize': 8})
plt.show()
```
