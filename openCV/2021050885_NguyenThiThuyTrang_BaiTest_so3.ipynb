# Yêu cầu 1:


### 1.1  Sinh viên đọc và hiển thị ảnh Pic/Pic_T1.png ở chế độ ảnh màu RGB


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
hoa = cv2.imread("Pic/Pic_T1.png",1)
plt.imshow(cv2.cvtColor(hoa, cv2.COLOR_BGR2RGB));
```

### 1.2. Lấy thông số của ảnh, thực hiện Hiển thị kích thước và số pixel điểm ảnh

```python
print("Kích thước ảnh: ",hoa.shape[0],'x',hoa.shape[1])
print("Số pixel điểm ảnh: ",hoa.size,'pixel')
```

### 1.3. Thay đổi kích thước ảnh trên về ảnh mới có kích thước (400x400)

```python
#Thay đổi kích thước ảnh về độ phân giải: 100 x 200
dsize = (400,400)

hoa_400x400 = cv2.resize(hoa,
                         dsize, 
                         interpolation=cv2.INTER_LINEAR) 

plt.figure(figsize=(10,5))

plt.imshow(cv2.cvtColor(hoa_400x400, cv2.COLOR_BGR2RGB))

plt.title('Ảnh thay đổi kích thước 400x400')

plt.show()

print("Ảnh sau khi thay đổi kích thước: ")
print("Kích thước ảnh: ",hoa_400x400.shape[0],'x',hoa_400x400.shape[1])
print("Số pixel điểm ảnh: ", hoa_400x400.size,'pixel')
```

### 1.4. Chuyển đổi ảnh màu RGB 400x400 ở trên sang ảnh Đen Trắng. Sau đó sử dụng phương thức Laplacian để xác định biên của bông hoa, thể hiện kết quả như hình minh họa. Lưu ảnh biên vào thư mục Pic đặt tên: MSV_Edge.jpg

```python
# Chuyển sang ảnh đen trắng
hoa_gray = cv2.cvtColor(hoa_400x400, cv2.COLOR_BGR2GRAY)
ret,hoa_binary = cv2.threshold(hoa_gray,100,255,cv2.THRESH_BINARY)
plt.imshow(hoa_binary,cmap='gray');


```

```python
hoa_Laplacian = cv2.Laplacian(hoa_binary,-1,7)
plt.imshow(hoa_Laplacian,cmap='gray');
```

```python
plt.figure(figsize=(10,8))
plt.subplot(1, 2, 1)
plt.imshow(hoa_binary, cmap='gray')
plt.title('Ảnh ở chế độ Đen - Trắng')

plt.subplot(1, 2, 2)
plt.imshow(hoa_Laplacian, cmap='gray')
plt.title('Đường biên của cánh hoa theo laplacian')
plt.show()
```

```python
cv2.imwrite('Pic/2021050885_Edge.jpg',hoa_Laplacian)
```

# Yêu cầu 2

### 2.1. Sinh viên đọc ảnh tầm soát ung thư Pic/Pic_T2.jpeg ở dạng ảnh xám và hiển thị như hình dưới đây:

```python
ungthu = cv2.imread("Pic/Pic_T2.jpeg",0)
plt.imshow(ungthu,cmap='gray');
```

### 2.2. Sử dụng phương pháp cắt ngưỡng ảnh – threshold với loại cắt ngưỡng (type) là THRESH_BINARY_INV và các giá trị ngưỡng phù hợp để thu được ảnh như sau:

```python
ret,ungthu_dao= cv2.threshold(ungthu,30,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize=(10,8))
plt.grid()
plt.imshow(ungthu_dao,cmap='gray');
```

### 2.3. Sử dụng phương pháp phát hiện biên Canny với các tham số phù hợp để phát hiện đường biên của ảnh như dưới đây:

```python
ungthu_Canny = cv2.Canny(ungthu_dao, 50,80,5)
plt.figure(figsize=(10,8))
plt.grid()
plt.imshow(ungthu_Canny,cmap='gray');
```

### 2.4. Hiển thị các kết quả thu được từ yêu cầu 1 đến 3 như dưới đây, lưu ảnh cắt ngưỡng và ảnh phát hiện cạnh vào thư mục Pic đặt tên lần lượt là MSV_threshold.png và MSV_canny.png

```python
# Hiển thị ảnh
plt.figure(figsize=(15,12))
plt.subplot(1,3,1)
plt.title("Ảnh gốc ban đầu")
plt.imshow(ungthu,cmap='gray')

plt.subplot(1,3,2)
plt.title("Ảnh cắt ngưỡng THRESH_BINARY_INV ")
plt.imshow(ungthu_dao,cmap='gray')

plt.subplot(1,3,3)
plt.title("Ảnh phát hiện biên với Canny")
plt.imshow(ungthu_Canny,cmap='gray')

plt.show()

```

```python
# Lưu ảnh
cv2.imwrite('Pic/2021050885_threshold.png',ungthu_dao)
cv2.imwrite("Pic/2021050885_canny.png",ungthu_Canny)
```
