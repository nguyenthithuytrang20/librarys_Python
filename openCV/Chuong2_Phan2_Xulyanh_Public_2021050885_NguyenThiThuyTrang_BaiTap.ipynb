## THỰC HÀNH 2.5: 
---
**Yêu cầu 1:** Sinh viên đọc ảnh images/Thuchanh2_5.jpg ở dạng xám

**Yêu cầu 2:** Thiết lập tham số c, a phù hợp để tăng cường chất lượng ảnh và hiển thị kết quả (slide). Lưu file ảnh đã  đổi vào thư mục Saves/MSV_gamma.jpg

<img src='images/Thuchanh2_5.jpg' width='300px'>

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import warnings
warnings.filterwarnings('ignore')
```

```python
#{SINH VIÊN VIẾT CODE Ở ĐÂY}
hand = cv2.imread("images/Thuchanh2_5.jpg",0)

def change_gama(img,c,a):
    
    img = c*pow(img,a)
    # chuyển về kiểu số nguyên
    img = img.astype(np.int32)
    return img

hand_gama = change_gama(hand,1,2.7)

plt.figure(figsize=(16,12))
plt.subplot(1,2,1)
plt.imshow(hand,cmap='gray')

plt.subplot(1,2,2)
plt.imshow(hand_gama,cmap='gray')

plt.show()



```

```python
cv2.imwrite('images/Saves/2021050885_gamma.jpg',hand_gama)
```

## THỰC HÀNH 2.6: 
---
**Yêu cầu 1:** Sinh viên đọc ảnh images/Thuchanh2_6.jpg ở dạng xám

**Yêu cầu 2:** Thực hiện cắt ngưỡng ảnh với (Type = cv2.THRESH_BINARY) lần lượt với T = 50, 100, 150. Hiển thị kết quả như slide minh hoạ. Lưu ảnh vào thư mục Saves

**Yêu cầu 3:** Thiết lập T = 75. Thực hiện cắt ngưỡng với Type = cv2.THRESH_TRUNC, cv2.THRESH_BINARY_INV, cv2.THRESH_TOZERO. Hiển thị kết quả như slide minh hoạ. Lưu ảnh vào thư mục Saves
<img src='images/Thuchanh2_6.jpg' width='300px'>

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#{SINH VIÊN VIẾT CODE Ở ĐÂY}
girl = cv2.imread('images/Thuchanh2_6.jpg',0)
girl.max()
```

```python
# Yêu cầu 1:
ret,girl1 = cv2.threshold(girl,50,255,cv2.THRESH_BINARY)
ret,girl2 = cv2.threshold(girl,100,255,cv2.THRESH_BINARY)
ret,girl3 = cv2.threshold(girl,150,255,cv2.THRESH_BINARY)
```

```python
# hiển thị ảnh
plt.figure(figsize=(12,8))
plt.subplot(1,4,1)
plt.title('Ảnh gốc ban đầu')
plt.imshow(girl,cmap='gray')

plt.subplot(1,4,2)
plt.title("Ảnh cắt ngưỡng T = 50")
plt.imshow(girl1,cmap='gray')

plt.subplot(1,4,3)
plt.title("Ảnh cắt ngưỡng T = 100")
plt.imshow(girl2,cmap='gray')

plt.subplot(1,4,4)
plt.title("Ảnh cắt ngưỡng T = 150")
plt.imshow(girl3,cmap='gray')
plt.show()
```

```python
cv2.imwrite('images/Saves/gilr.jpg',girl)
cv2.imwrite('images/Saves/gilr1.jpg',girl1)
cv2.imwrite('images/Saves/gilr2.jpg',girl2)
cv2.imwrite('images/Saves/gilr3.jpg',girl3)
```

```python
#Yêu cầu 2:
T=75
ret,girl2_1=cv2.threshold(girl,T,255,cv2.THRESH_TRUNC)
ret,girl2_2=cv2.threshold(girl,T,255,cv2.THRESH_BINARY_INV)
ret,girl2_3=cv2.threshold(girl,T,255,cv2.THRESH_TOZERO)
```

```python
# hiển thị ảnh
plt.figure(figsize=(16,14))
plt.subplot(1,4,1)
plt.title("Ảnh gốc ban đầu")
plt.imshow(girl,cmap='gray')

plt.subplot(1,4,2)
plt.title("T=75&THRESH_TRUNC")
plt.imshow(girl2_1,cmap='gray')

plt.subplot(1,4,3)
plt.title("T=75&THRESH_BINARY_INV")
plt.imshow(girl2_2,cmap='gray')

plt.subplot(1,4,4)
plt.title("T=75&THRESH_TOZERO")
plt.imshow(girl2_3,cmap='gray')
plt.show()
```

```python
cv2.imwrite('images/Saves/girl_TRUNC.jpg',girl2_2)
cv2.imwrite('images/Saves/girl_BINARY__INV.jpg',girl2)
```

## THỰC HÀNH 2.7: 
---
Yêu cầu :
* Yêu cầu 1: Sinh viên đọc ảnh images/Thuchanh2_6.jpg ở dạng ảnh xám. Thực hiện cân bằng sáng trên ảnh Xám, hiển thị ảnh gốc và ảnh sau cần bằng cùng với Histogram tương ứng của mỗi ảnh.
* Yêu cầu 2: Sinh viên đọc ảnh images/Thuchanh2_6.jpg ở dạng ảnh màu. Thực hiện cân bằng sáng trên ảnh màu, hiển thị ảnh gốc và ảnh sau cần bằng ở dạng ảnh màu.
* Yêu cầu 3: Lưu ảnh xám và ảnh màu đã cân bằng vào thư mục images/Saves.

<img src = 'images/Thuchanh2_6.jpg' width='150px'>

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_girl_gray = cv2.imread('images/Thuchanh2_6.jpg',0)
print(img_girl_gray.max())
```

```python
#YÊU CẦU 1: Đọc ảnh xám:
img_girl_gray = cv2.imread('images/Thuchanh2_6.jpg',0)

#Thực hiện cân bằng sáng với phương thức equalizeHist():
img_girl_equalizeHist = cv2.equalizeHist(img_girl_gray)

#Hiển thị ảnh và Histogram tương ứng.
#Ảnh 1
plt.figure(figsize=(10,5))
plt.figure(figsize=(12,6))
plt.subplot(2, 2, 1)
plt.imshow(img_girl_gray, cmap='gray')
plt.title('Ảnh Gốc')

plt.subplot(2,2,3)
plt.title('Histogram của ảnh gốc')
plt.hist(img_girl_gray.ravel(),256,[0,256])

plt.subplot(2, 2, 2)
plt.imshow(img_girl_equalizeHist, cmap='gray')
plt.title('Ảnh đã cân bằng sáng ')

plt.subplot(2,2,4)
plt.title("Histogram của ảnh đã cân bằng sáng")
plt.hist(img_girl_equalizeHist.ravel(),256,[0,256])
plt.show()
```

```python
#Đọc ảnh ở dạng ảnh màu
img_girl_color = cv2.imread('images/Thuchanh2_6.jpg',1)

#Chuyển ảnh về dạng YUV
img_girl_YUV = cv2.cvtColor(img_girl_color,cv2.COLOR_BGR2YUV)

#Cân bằng sáng trên ảnh YUV
img_girl_YUV[:,:,0] = cv2.equalizeHist(img_girl_YUV[:,:,0])

#Chuyển ảnh sau khi cân bằng sáng về hệ màu RGB
img_girl_equalizeHist_color = cv2.cvtColor(img_girl_YUV,cv2.COLOR_YUV2BGR)

#Hiển thị
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title("Ảnh màu gốc")
plt.imshow(cv2.cvtColor(img_girl_color,cv2.COLOR_BGR2RGB))

plt.subplot(1,2,2)
plt.title("Ảnh màu đã cân bằng")
plt.imshow(cv2.cvtColor(img_girl_equalizeHist_color,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#Lưu
cv2.imwrite('images/Saves/img_gray_histogram.jpg',img_girl_equalizeHist)
cv2.imwrite('images/Saves/img_color_histogram.jpg',img_girl_equalizeHist_color)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Đọc ảnh Thuchanh2_6.jpg ở dạng ảnh Xám.
#Thực hiện cân bằng sáng toàn ảnh, cân bằng theo từng vùng 15x15.
#Hiển thị 3 ảnh: Ảnh gốc - Ảnh cân bằng toàn ảnh - Ảnh cân bằng theo vùng 15x15.

img = cv2.imread('images/Thuchanh2_6.jpg',0)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
#Cân bằng sáng toàn ảnh
img_equalizeHist = cv2.equalizeHist(img)

#Cân bằng sáng theo vùng
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(15,15))
img_clahe = clahe.apply(img)

# hiển thị ảnh
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.title("Ảnh gốc")
plt.imshow(img,cmap='gray')

plt.subplot(1,3,2)
plt.title('Ảnh cân bằng toàn ảnh')
plt.imshow(img_equalizeHist,cmap='gray')

plt.subplot(1,3,3)
plt.title("Ảnh cân bằng sáng theo vùng")
plt.imshow(img_clahe,cmap='gray')

plt.show()
```
