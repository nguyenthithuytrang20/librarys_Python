## THỰC HÀNH 1.5: 
---
Yêu cầu: 
1. Sinh viên đọc ảnh màu images/Thuchanh_5.jpg
2. Cho biết chiều cao, chiều rộng và số pixel điểm ảnh.
3. Tách thành các kênh màu B-G-R tương ứng và hiển thị như yêu cầu trong slide
4. Ghi từng kênh màu thành các ảnh riêng vào thư mục images/Saves đặt tên như sau: MSV_blue.jpg; MSV_green.jpg; MSV_red.jpg

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
part = 'images/Thuchanh_5.jpg'
#1. Đọc ảnh với chế độ mặc định: 
part_2 = cv2.imread(part)
plane = cv2.cvtColor(part_2,cv2.COLOR_BGR2RGB)
plt.title("PLANE_RGB")
plt.imshow(plane)
plt.show()



```

```python
#2. Hiển thị thông tin ảnh:
print('Chiều cao của ảnh là:', plane.shape[0])
print('Chiều rộng của ảnh là:',plane.shape[1])
print('Số pixel điểm ảnh:',plane.size)

```

```python
#3. Tách các kênh màu B, G, R từ ảnh màu
b_channel = None
g_channel = None
r_channel = None

# Lấy dữ liệu các kênh màu: BGR 
b_channel = plane.copy()
b_channel[:, :, [1,2]] = 0 

g_channel = plane.copy()
g_channel[:, :, [0, 2]] = 0 

r_channel = plane.copy()
r_channel[:, :, [0, 1]] = 0 
```

```python
#Hiển thị ảnh gốc và các kênh màu RGB với Matplotlib

plt.figure(figsize=(12,8))
plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(part_2, cv2.COLOR_BGR2RGB))
plt.grid()
plt.title('Ảnh RGB')

plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(b_channel, cv2.COLOR_BGR2RGB))
plt.title('Blue Channel')

plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(g_channel, cv2.COLOR_BGR2RGB))
plt.title('Green Channel')

plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(r_channel, cv2.COLOR_BGR2RGB))
plt.title('Red Channel')
plt.show()

```

```python
#4.Ghi ảnh các kênh màu
cv2.imwrite('images/Saves/2021050885_blue.jpg',b_channel)
cv2.imwrite('images/Saves/2021050885_green.jpg',g_channel)
cv2.imwrite('images/Saves/2021050885_red.jpg',r_channel)
```

# 1. Hệ màu HSV
##  HSV được đặt tên như vậy cho ba giá trị
* Hue(màu sắc): là phần màu của mô hình màu và được biểu thị dưới dạng một số từi 0 ->360
* Saturation(độ bão hòa): Là lượng màu xám trong màu, từ 0->100 phần trăm. Giảm độ bão hòa để có nhiều màu xám hơn
* Value(giá trị hoặc độ sáng):Mo tả dộ sáng, cường độ củ màu sắc, 0->100%; 0 là hoàn toàn đen; 100 là sáng nhất cho thấy rõ màu nhất.

##  HSV được sử dụng như thế nào?
* Không gian màu HSV được sử dụng khi chọn màu cho sơn hoặc mực vì HSV thể hiện rõ hơn cách mọi người liên quan đến màu sắc hơn không gian màu RGB.
* Bánh xe màu HSV cũng được sử dụng để tạo ra đồ họa chất lượng cao. Mặc dù ít được biết đến hơn so với người anh em họ RGB và CMYK, phương pháp HSV có sẵn trong nhiều chương trình phần mềm chỉnh sửa hình ảnh cao cấp
* Chọn màu HSV bắt đầu bằng cách chọn một trong các màu có sẵn, đó là cách hầu hết mọi người liên quan đến màu sắc, sau đó điều chỉnh giá trị độ sáng và độ bóng

```python
img_RGB=cv2.imread('images/Tulips/Pic_0.jpg',1)
img_bgr = cv2.imread('images/pic_0.jpg',1)
```

```python
# Hệ màu HSV
fig,ax = plt.subplots(1,2, figsize=(12,6))
ax[0].imshow(cv2.cvtColor(img_RGB,cv2.COLOR_BGR2RGB))
ax[0].set_title('tulips_RGB')

ax[1].imshow(cv2.cvtColor(img_RGB,cv2.COLOR_RGB2HSV))
ax[1].set_title('tulips_HSV')
plt.show()
```

# 2. Hệ màu CMYK
### CMYK là từ viết tắt tiếng Anh của cơ chế hệ màu trừ, thường được sử dụng trong in ấn. Nó bao gồm các màu sau:
* C = Cyan (xanh)

* M = Magenta (hồng)

* Y = Yellow (vàng)

* K = Black (Đen) (sở dĩ dùng từ K để chỉ màu đen vì ký tự B đã được dùng để chỉ màu Blue, ngoài ra K còn có nghĩa là Key, mang ý chỉ cái gì đó là chủ yếu, là then chốt)

### Nguyên lý làm việc
*  CMYK là hấp thụ ánh sáng. Màu mà ta nhìn thấy là từ phần của ánh sáng không bị hấp thụ, hay nói cách khác, chúng hoạt động trên cơ chế những vật không tự phát ra ánh sáng mà chỉ phản xạ ánh sáng từ các nguồn khác chiếu tới

* CMYK sẽ loại trừ ánh sáng đi từ ánh sáng gốc là màu trắng để tạo ra các màu sắc khác. 3 màu Cyan, Magenta và Yellow khi kết hợp sẽ tạo ra một màu đen

* CMYK thường được sử dụng khi thiết kế phục vụ cho mục đích in ấn các thiết kế như poster, brochure, name card, catalogue, sách hoặc tạp chí,…
